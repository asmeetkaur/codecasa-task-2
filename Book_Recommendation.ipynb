```python
def recommend(book_name):
    # Check if the book name exists in the index
    if book_name in pt.index:
        index = np.where(pt.index == book_name)[0][0]

        # Check if the index is within the bounds of similarity_score
        if index < len(similarity_score):
            similar_items = sorted(enumerate(similarity_score[index]), key=lambda x: x[1], reverse=True)[1:11]

            for i in similar_items:
                print(pt.index[i[0]])
        else:
            print(f"Index '{index}' is out of bounds.")
    else:
        print(f"The book '{book_name}' was not found in the dataset.")

# Example usage
recommend('A Bend in the Road')
```
